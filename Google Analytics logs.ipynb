```python
def get_anomalies_by_users_a_day(df, mean, fluct=0.25, col='normal'):
    if not isinstance(fluct, float) or fluct < 0 or fluct > 1:
        print("Give fluctuation in range of [0-1]")
        return
    for index, row in df.iterrows():
        #value = float(row[col].replace(u'\xa0',''))
        value = row[col]
        if value > mean*(1 + fluct) or value < mean*(1 - fluct):
            print("Day of anomaly: {}".format(row['day']))
```

```python
get_anomalies_by_users_a_day(df, mean, 0.4)
```
